## Arbitrary Arguments (*args)

```python
# it give output in tuple and

def add(* args):
  s = 0
  for x in args:
    s = s + x
  return s

result = add(1, 2, 3, 5)
print(result)
```

## Arbitrary Keyword Arguments (**kwargs)

```python
def addr(**kwargs):
  for k, v in kwargs.items():
    print("{}:{}".format(k, v))

print("pass two keyword args")
addr(Name="john", city="Mumbai")
print("pass four keyword args")
```

## Comprehension ##

Comprehensions are concise and readable ways to create lists, sets, or dictionaries from existing iterables (like lists, tuples, strings, or ranges).
They replace loops like for or map() with a single expressive line.

```python
# List comprehension : A list comprehension lets you build a new list
# by applying an expression to each element of
# an iterable — optionally including a condition.
# Syntax: [expression for item in iterable if condition]

squares = [x ** 2 for x in range(1, 6) if x % 2 == 0]
print(squares)
```

```python
# Dictionary Comprehension : Creates a dictionary in one line using key–value expressions.
# Syntax: {key_expr: value_expr for item in iterable if condition}

squares_dict = {x : x**2 for x in range(1, 6) if x % 2 == 0}
print(squares_dict)
```

```python
# Set Comprehension : A set comprehension builds a set (collection of unique items) from an iterable.
# Syntax: {expression for item in iterable if condition}
# Automatically removes duplicates (since it’s a set).

squares_set = {x ** 2 for x in range(1, 6)}
print(squares_set)
```

***Nested Compreshension*** : When a comprehension contains another comprehension (like nested loops).

```python
matrix = [[1, 2, 3], [4, 5, 6]]
flattened = [num for row in matrix for num in row]

print(flattened)
```

# ***` Data Structure (collections) of python `***

```python
# list :
# Mutable ordered sequence of elements.
# Can store mixed data types (int, str, list, etc.).
# Defined with square brackets: []

# Creating List
numbers = [1, 2, 3, 4, 5]
mixed = [1, "Anoj", 3.5, True]
empty = []

print(numbers)
print(mixed)

print()
print("---------Accessing Element------------")

# Accessing Element
print(numbers[0])
print(numbers[-1])

print()
print("---------slicing------------")

# slicing

# : [start(deflt = 0) : end(deflt = last) : step(deflt: 1)]
print(numbers[1:3])
print(numbers[:3])
print(numbers[2:])
print(numbers[1::2])


print()
print("---------other List method------------")
# other List method

numbers.append(6) # add value in list at the end
numbers.extend([7, 8]) # Add multiple values in the list
numbers.insert(4, 33) # insert value in the index
# numbers.remove(5) # remove the value that occur in the first occurrence
numbers.pop(3) # removes the index values and return elements
print(numbers)

numbers.sort() # sort
print(numbers)
# numbers.reverse()

```

```python
# Tuple :
# Immutable ordered sequence — cannot be changed after creation.
# Defined with parentheses ().

p = ("anoj", 22, "nepal", 22)

# unpacking tuple
name, age, country, ages = p
print(name, age, country, ages)

print()
print("---------other Tuple method------------")

p_count = p.count(22) # count the occurrences of the passed value
print(p_count)

p_index = p.index(22) # it find the index of the passed value

```

```python
# Set :
# Unordered, mutable collection of unique elements.
# Defined with curly braces {} or set().

# No duplicates allowed.
# Sets are unordered → no indexing or slicing.

nums = {1, 2, 3, 4, 6}
empty = set()

print()
print("---------set------------")

nums.add(5) # add the unique values
nums.remove(4) # remove the passed value
nums.discard(6) # remove the value (no error if not found)
nums.pop() # remove and return a random value
print(nums)
nums.clear() # remove all the
print(nums)
```

```python
# Dictionaries :
# Key–value pairs collection.
# Keys must be unique and immutable (e.g., str, int, tuple).
# Defined with curly braces {key: value}.

students = {
    "name" : "Anoj",
    "age" : 21,
    "country" : "Nepal"
}

print()
print("---------Dictionaries------------")
# Access, Add, Update, Delete
print(students["name"]) # Access
students["age"] = 22 # update
students["email"] = "anoj@gmail.com" # add
del students["country"] # delete

print(students)

print()
print("---------other Dic method------------")

key = students.keys() # return all the keys
print ( f"keys : {key}")
value = students.values() # return all the values
print( f"values : {value}")

items = students.items() # return all the items (key, value)
print (f"items : {items}")

get = students.get("name") # get safe key access if there is no value it return none
print(f"get : {get}")


stud2 = {
    "name" : "ram",
    "age" : 23
}
students.update(stud2) # it update the dic or merge the dic if the key, value already exit it update values of second dic
print(students)


```
