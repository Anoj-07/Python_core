## Arbitrary Arguments (*args)

```python
# it give output in tuple and

def add(* args):
  s = 0
  for x in args:
    s = s + x
  return s

result = add(1, 2, 3, 5)
print(result)
```

## Arbitrary Keyword Arguments (**kwargs)

```python
def addr(**kwargs):
  for k, v in kwargs.items():
    print("{}:{}".format(k, v))

print("pass two keyword args")
addr(Name="john", city="Mumbai")
print("pass four keyword args")
```

## Comprehension ##

Comprehensions are concise and readable ways to create lists, sets, or dictionaries from existing iterables (like lists, tuples, strings, or ranges).
They replace loops like for or map() with a single expressive line.

```python
# List comprehension : A list comprehension lets you build a new list
# by applying an expression to each element of
# an iterable — optionally including a condition.
# Syntax: [expression for item in iterable if condition]

squares = [x ** 2 for x in range(1, 6) if x % 2 == 0]
print(squares)
```

```python
# Dictionary Comprehension : Creates a dictionary in one line using key–value expressions.
# Syntax: {key_expr: value_expr for item in iterable if condition}

squares_dict = {x : x**2 for x in range(1, 6) if x % 2 == 0}
print(squares_dict)
```

```python
# Set Comprehension : A set comprehension builds a set (collection of unique items) from an iterable.
# Syntax: {expression for item in iterable if condition}
# Automatically removes duplicates (since it’s a set).

squares_set = {x ** 2 for x in range(1, 6)}
print(squares_set)
```

***Nested Compreshension*** : When a comprehension contains another comprehension (like nested loops).

```python
matrix = [[1, 2, 3], [4, 5, 6]]
flattened = [num for row in matrix for num in row]

print(flattened)
```
